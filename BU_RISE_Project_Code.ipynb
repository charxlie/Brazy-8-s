## Binary Classification

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
# Create classifiers
lr = LogisticRegression()
svc = LinearSVC(C = 1.0)
rfc = RandomForestClassifier(n_estimators = 100)
```

```python
train, test = train_test_split(df, test_size = 0.2)
```

```python
train.head()
```

```python
train_feat = train.iloc[:,:11]  # independent vars
train_targ = train['MSval']
```

```python
test_feat = test.iloc[:,:11]
test_targ = test['MSval']
```

### Logistic Regression
2nd best

```python
lr.fit(train_feat, train_targ)
```

```python
lr.score(train_feat, train_targ)
```

```python
lr.score(test_feat, test_targ)
```

```python
lr.coef_  # log odds / logits / odds ratio - check these out to get function
np.transpose(lr.coef_)
```

```python
preds = lr.predict(test_feat)
```

```python
pd.crosstab(preds, test_targ)
```

```python
confusion_matrix(lr.predict(train_feat), train_targ)
```

```python
confusion_matrix(lr.predict(test_feat), test_targ)
```
